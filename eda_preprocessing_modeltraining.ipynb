 ## **2. Feature Engineering and Selection:**

## **3. Model Selection and Training:**

```python
#preparing dataframes for training
X = df.drop('reduced_life_expectancy',axis=1)
y = df['reduced_life_expectancy']

#defing train-test split
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(X, y, test_size=0.25, random_state=2)
```

```python
# Importing regression models to be tested
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor

# Defining function to create list of models to be tested
def create_models(rs=42):
    models = []
    models.append(('Decision Tree Regressor', DecisionTreeRegressor(random_state=rs)))
    models.append(('K Neighbours Regressor', KNeighborsRegressor()))
    models.append(('Linear Regression', LinearRegression()))
    models.append(('Random Forest Regressor', RandomForestRegressor(random_state=rs)))
    return models

models = create_models()
```

```python
from sklearn.metrics import mean_squared_error, r2_score

# Evaluating each model
print("Performance metrics used are: MSE, RMSE, and Coefficient of Determination")
for name, model in models:
  print("\n-------------------------------")
  print('{}\n-------------------------------'.format(name))
  model.fit(xtrain,ytrain)

  # training
  ytrain_pred = model.predict(xtrain)
  mse = mean_squared_error(ytrain, ytrain_pred)
  rmse = (np.sqrt(mean_squared_error(ytrain, ytrain_pred)))
  r2 = r2_score(ytrain, ytrain_pred)
  print("Training Performance")
  print('MSE:{}'.format(mse))
  print('RMSE:{}'.format(rmse))
  print('R2:{}'.format(r2))

  # test
  ytest_pred = model.predict(xtest)
  mse = mean_squared_error(ytest, ytest_pred)
  rmse = (np.sqrt(mean_squared_error(ytest, ytest_pred)))
  r2 = r2_score(ytest, ytest_pred)

  print("Test Performance")
  print('MSE:{}'.format(mse))
  print('RMSE:{}'.format(rmse))
  print('R2:{}'.format(r2))
```

Decision Tree Regressor achieves perfect training performance but shows some overfitting with a lower testing R2 value of around 0.98

K Neighbours Regressor has a decent fit, but comparatively lowest R2 value of around 0.73

Linear Regression gives a coeff of determination around 0.75

Random Forest appears to have the best performance with the lowest test MSE, RMSE values and the highest test R2 value however it may be overfitting at 0.99+

## **4. Model Improvement:**

Since the Random Forest Regressor showed the best performance out of all models, we will select it for further improvement. <br>
Instead of performing a computationally expensive grid search on all important hyperparameter combinations, which would result in **1600 combinations** in our case, first we will conduct random search to narrow down the ranges for each hyperparameter values. Next, we will perform grid search only on the most promising hyperparameter ranges for a more efficient and focused tuning process.

```python
from sklearn.model_selection import RandomizedSearchCV

# define the hyperparameter search space
param_dist = {
    'n_estimators': [int(x) for x in np.linspace(start = 200, stop = 2000, num = 10)],
    'max_features': ['auto', 'sqrt'],
    'max_depth': [int(x) for x in np.linspace(start = 10, stop = 100, num = 10)],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2],
    'bootstrap': [True, False]
}

# create the model
rf1 = RandomForestRegressor(random_state=42)

# define the random search
n_iter_search = 50
random_search = RandomizedSearchCV(estimator=rf1, param_distributions=param_dist, n_iter=n_iter_search, cv=5, n_jobs=-1)

# perform the random search
random_search.fit(xtrain, ytrain)

# print the best parameters and the corresponding performance metrics
print("Best parameters: ", random_search.best_params_)

ytrain_pred = random_search.predict(xtrain)
print("\nTraining Performance After Hyperparameter Tuning")
print("MSE:", mean_squared_error(ytrain, ytrain_pred))
print("RMSE:", np.sqrt(mean_squared_error(ytrain, ytrain_pred)))
print("R2:", r2_score(ytrain, ytrain_pred))

ytest_pred = random_search.predict(xtest)
print("\nTest Performance After Hyperparameter Tuning")
print("MSE:", mean_squared_error(ytest, ytest_pred))
print("RMSE:", np.sqrt(mean_squared_error(ytest, ytest_pred)))
print("R2:", r2_score(ytest, ytest_pred))
```

Random search resulted in training R2 score of **100%**, which raises concerns about overfitting. We will address this by performing grid search to explore the optimal values of `n_estimators` and `max_depth`, ranging around **2000** and **50**, respectively for pruning the model to avoid overfitting.

```python
from sklearn.model_selection import GridSearchCV

# define the hyperparameter search space
param_grid = {
    'n_estimators': [2000, 2100, 2200],
    'max_features': ['sqrt'],
    'max_depth': [35, 40, 50],
    'bootstrap': [False]
}

# create the model
rf2 = RandomForestRegressor(random_state = 42)

# define the grid search
grid_search = GridSearchCV(estimator=rf2, param_grid=param_grid, cv=5, n_jobs=-1)

# perform the grid search
grid_search.fit(xtrain, ytrain)

# print the best parameters and the corresponding performance metrics
print("Best parameters: ", grid_search.best_params_)

ytrain_pred = grid_search.predict(xtrain)
print("\nTraining Performance After Hyperparameter Tuning")
print("MSE:", mean_squared_error(ytrain, ytrain_pred))
print("RMSE:", np.sqrt(mean_squared_error(ytrain, ytrain_pred)))
print("R2:", r2_score(ytrain, ytrain_pred))

ytest_pred = grid_search.predict(xtest)
print("\nTest Performance After Hyperparameter Tuning")
print("MSE:", mean_squared_error(ytest, ytest_pred))
print("RMSE:", np.sqrt(mean_squared_error(ytest, ytest_pred)))
print("R2:", r2_score(ytest, ytest_pred))
```

Through grid search, we decreased the training R2 score by 0.1% and increasing the training MSE and RMSE while keeping test performance stable compared to random search model, indicating increased generalization. Notably, the difference between the training and test R2 score in improved model is a mere 0.004, which suggests that overfitting is not an issue.<br>
Our improved model also outperformed the base model's R2 score by 0.22%, and reduced the MSE and RMSE scores by 1.02% and 3.19% respectively.
